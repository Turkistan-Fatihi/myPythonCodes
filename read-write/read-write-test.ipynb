```python
def menu():
    print("\n=== DOSYA İŞLEMLERİ MENÜSÜ ===")
    print("1 - Dosya Oku")
    print("2 - Dosyaya Yaz")
    print("3 - Dosya İçeriğini Sil")
    print("4 - Çıkış")
    secim = input("Lütfen bir işlem seçin (1/2/3/4): ")
    return secim

def dosya_oku(dosya_adi):
    try:
        with open(dosya_adi, "r") as dosya:
            print("\n--- Dosya İçeriği ---")
            print(dosya.read())
    except FileNotFoundError:
        print("Dosya bulunamadı.")

def dosyaya_yaz(dosya_adi):
    veri = input("Yazmak istediğiniz veriyi girin: ")
    with open(dosya_adi, "a") as dosya:
        dosya.write("\n" + veri)
    print("Veri başarıyla dosyaya eklendi.")

def dosya_sil(dosya_adi):
    with open(dosya_adi, "w") as dosya:
        pass  # içeriği temizler
    print("Dosya içeriği silindi.")

# Ana program döngüsü
dosya_adi = "pythontest.txt"

while True:
    secim = menu()
    
    if secim == "1":
        dosya_oku(dosya_adi)
    elif secim == "2":
        dosyaya_yaz(dosya_adi)
    elif secim == "3":
        dosya_sil(dosya_adi)
    elif secim == "4":
        print("Programdan çıkılıyor...")
        break
    else:
        print("Geçersiz seçim. Lütfen 1, 2, 3 veya 4 girin.")
```
